```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the model
model = GPT2LMHeadModel.from_pretrained("/content/drive/MyDrive/CS271_Project/mySavedModel")

# Load the tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("/content/drive/MyDrive/CS271_Project/mySavedModel")
```

```python
from transformers import pipeline

# Create a text generation pipeline
pl = pipeline("text-generation", model="/content/drive/MyDrive/CS271_Project/mySavedModel")
```

```python
def create_prompt(ingredients):
    ingredients = ','.join([x.strip().lower() for x in ingredients.split(',')])
    ingredients = ingredients.strip().replace(',','\n')
    s = f"<|startoftext|>Ingredients:\n{ingredients}\n"
    return s
```

```python
import pandas as pd

My_data = pd.read_csv('/content/drive/MyDrive/CS271_Project/recipes_w_search_terms.csv')
My_data = My_data.sample(frac=1)
My_data.reset_index(drop=True,inplace=True)
My_data.info(verbose = False, memory_usage = "deep")
filtered_df = My_data[My_data['tags'].apply(lambda x: '30-minutes-or-less' in x)]

print(filtered_df.shape[0])
df = filtered_df[["ingredients_raw_str", "steps"]]
df.info(verbose = False, memory_usage = "deep")
df.drop_duplicates(inplace=True)
df.shape[0]
df.head()
```

```python
recipes = []
for index, row in df.iterrows():
    ingredients = eval(row['ingredients_raw_str'])
    steps = eval(row['steps'])
    recipe_str = "Ingredients:\n" + "\n".join(ingredients) + "\n\nSteps:\n" + "\n".join(steps)
    recipes.append(recipe_str)

```

```python
#Implementing RAG with guardrails
from sentence_transformers import SentenceTransformer, util
import torch
import faiss

# Load a model for sentence embeddings
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Assume `recipes` is a list of recipe texts
# Need to prepare recipes list

recipe_embeddings = embedding_model.encode(recipes, convert_to_tensor=True)

# Build a FAISS index
index = faiss.IndexFlatL2(recipe_embeddings.shape[1])
index.add(recipe_embeddings.cpu().numpy())


```

```python
def retrieve_recipes(query, index, recipe_embeddings, recipes, embedding_model):
    # query_embedding = embedding_model.encode(query, convert_to_tensor=True)
    # _, indices = index.search(query_embedding.cpu().numpy(), k=5)
    # return [recipes[idx] for idx in indices[0]]
        # Generate embedding for the query
    query_embedding = embedding_model.encode(query, convert_to_tensor=True)
    # Reshape the query embedding to have one row and multiple columns
    query_embedding = query_embedding.unsqueeze(0)

    # Search in the FAISS index
    _, indices = index.search(query_embedding.cpu().numpy(), k=5)

    # Fetch the retrieved recipes
    return [recipes[idx] for idx in indices[0]]
```

```python
#Importing fine-tuned GPT2 model
# from transformers import GPT2LMHeadModel, GPT2Tokenizer

# # Load your fine-tuned GPT-2 model and tokenizer
# model = GPT2LMHeadModel.from_pretrained("path/to/your/model")
# tokenizer = GPT2Tokenizer.from_pretrained("path/to/your/model")

#Setting up Retrieval-Augmented Generation (RAG)
def generate_recipe(query):
    retrieved_recipes = retrieve_recipes(query, index, recipe_embeddings, recipes, embedding_model)
    input_text = query + "\n" + "\n".join(retrieved_recipes)
    input_ids = tokenizer.encode(input_text, return_tensors='pt')

    # Increase max_length
    output = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Example usage
output = generate_recipe("chicken, rice, broccoli")
print(output)


```

```python
#NEW
def generate_recipe(query):
    retrieved_recipes = retrieve_recipes(query, index, recipe_embeddings, recipes, embedding_model)
    input_text = query + "\n" + "\n".join(retrieved_recipes)
    input_ids = tokenizer.encode(input_text, return_tensors='pt')
    output = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return {"recipe": generated_text}



import guardrails as gr

# Define guardrails
guard = gr.Guard.from_rail_string("""
<rail version="0.1">

<output>
    <string name="recipe" description="A generated recipe text" />
</output>

<prompt>
    Generate a recipe using the following ingredients: {{ ingredients }}.
    <output/>
</prompt>

</rail>
""")

# Example usage
output = generate_recipe("chicken, rice, broccoli")

# Validate Output
output_with_guardrails = guard.validate(output)

if output_with_guardrails.error:
    print(f"Error: {output_with_guardrails.error}")
else:
    print(output_with_guardrails.validated_output["recipe"])

```
